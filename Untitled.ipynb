```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn
import matplotlib.dates as md
from matplotlib import pyplot as plt
from sklearn import preprocessing
from keras.layers.core import Dense, Activation, Dropout
from keras.layers import LSTM
from keras.models import Sequential
import time
```

```python
dp=pd.read_csv('C:\\Users\\MY PC\\OneDrive\\Desktop\\NASA_Bearing_Data\\archive\\realKnownCause\\realKnownCause\\ambient_temperature_system_failure.csv')
dp.head()
```

```python
figsize=(10,5)
dp.plot(x='timestamp', y='value', figsize=figsize, title='Temperature (Degree Fahrenheit)');
plt.grid();
plt.show();
```

```python
dp['timestamp'] = pd.to_datetime(dp['timestamp'])
dp['value'] = (dp['value'] - 32) * 5/9
dp.plot(x='timestamp', y='value', figsize=figsize);
plt.title('Temperature (Degree Celcius)', fontsize=16);
plt.grid();
plt.show();
```

```python
dp['hours'] = dp['timestamp'].dt.hour
dp['daylight'] = ((dp['hours'] >= 7) & (dp['hours'] <= 22)).astype(int)
dp['DayOfTheWeek'] = dp['timestamp'].dt.dayofweek
dp['WeekDay'] = (dp['DayOfTheWeek'] < 5).astype(int)
```

```python
outliers_fraction = 0.01

dp['time_epoch'] = (dp['timestamp'].astype(np.int64)/100000000000).astype(np.int64)

dp['categories'] = dp['WeekDay']*2 + dp['daylight']
a = dp.loc[dp['categories'] == 0, 'value']
b = dp.loc[dp['categories'] == 1, 'value']
c = dp.loc[dp['categories'] == 2, 'value']
d = dp.loc[dp['categories'] == 3, 'value']
```

```python
figsize=(10,5)
fig, ax = plt.subplots(figsize=figsize)
a_heights, a_bins = np.histogram(a)
b_heights, b_bins = np.histogram(b, bins=a_bins)
c_heights, c_bins = np.histogram(c, bins=a_bins)
d_heights, d_bins = np.histogram(d, bins=a_bins)
width = (a_bins[1] - a_bins[0])/6
ax.bar(a_bins[:-1], a_heights*100/a.count(), width=width, facecolor='blue', label='Weekend Night')
ax.bar(b_bins[:-1]+width, (b_heights*100/b.count()), width=width, facecolor='green', label ='Weekend Light')
ax.bar(c_bins[:-1]+width*2, (c_heights*100/c.count()), width=width, facecolor='red', label ='Weekday Night')
ax.bar(d_bins[:-1]+width*3, (d_heights*100/d.count()), width=width, facecolor='black', label ='Weekday Light')
plt.legend()
plt.show()
```

```python
data_n = dp[['value', 'hours', 'daylight', 'DayOfTheWeek', 'WeekDay']]
min_max_scaler = preprocessing.StandardScaler()
np_scaled = min_max_scaler.fit_transform(data_n)
data_n = pd.DataFrame(np_scaled)
```

```python
prediction_time = 1 
testdatasize = 1000
unroll_length = 50
testdatacut = testdatasize + unroll_length  + 1
```

```python
x_train = data_n[0:-prediction_time-testdatacut].values
y_train = data_n[prediction_time:-testdatacut  ][0].values
```

```python
x_test = data_n[0-testdatacut:-prediction_time].values
y_test = data_n[prediction_time-testdatacut:  ][0].values
```

```python
def unroll(data,sequence_length=24):
    result = []
    for index in range(len(data) - sequence_length):
        result.append(data[index: index + sequence_length])
    return np.asarray(result)
```

```python
x_train = unroll(x_train,unroll_length)
x_test  = unroll(x_test,unroll_length)
y_train = y_train[-x_train.shape[0]:]
y_test  = y_test[-x_test.shape[0]:]
```

```python
print("x_train", x_train.shape)
print("y_train", y_train.shape)
print("x_test", x_test.shape)
print("y_test", y_test.shape)
```

```python
model = Sequential()
model.add(LSTM(units=50,input_dim=x_train.shape[-1],return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(100, return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units=1))
model.add(Activation('linear'))

start = time.time()
model.compile(loss='mse', optimizer='rmsprop')
print('compilation time : {}'.format(time.time() - start))
```

```python
model.fit(x_train, y_train, batch_size=3028, epochs=50, validation_split=0.1)
```

```python
plt.figure(figsize = (10, 5))
plt.plot(model.history.history['loss'], label = 'Loss')
plt.plot(model.history.history['val_loss'], label = 'Val_Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
plt.legend()
```

```python
loaded_model = model
diff=[]
ratio=[]
p = loaded_model.predict(x_test)
for u in range(len(y_test)):
    pr = p[u][0]
    ratio.append((y_test[u]/pr)-1)
    diff.append(abs(y_test[u]- pr))
```

```python
plt.figure(figsize = (10, 5))
plt.plot(p,color='red', label='Prediction')
plt.plot(y_test,color='blue', label='Test Data')
plt.legend(loc='upper left')
plt.grid()
plt.legend()
```

```python
diff = pd.Series(diff)
number_of_outliers = int(outliers_fraction*len(diff))
threshold = diff.nlargest(number_of_outliers).min()
test = (diff >= threshold).astype(int)
complement = pd.Series(0, index=np.arange(len(data_n)-testdatasize))
dp['anomaly27'] = complement.append(test, ignore_index='True')
print(dp['anomaly27'].value_counts())
```

```python
plt.figure(figsize=(10,5))
a = dp.loc[dp['anomaly27'] == 1, ['time_epoch', 'value']] #anomaly
plt.plot(dp['time_epoch'], dp['value'], color='blue')
plt.scatter(a['time_epoch'],a['value'], color='red', label = 'Anomaly')
plt.axis([1.370*1e7, 1.405*1e7, 15,30])
plt.grid()
plt.legend()
```
